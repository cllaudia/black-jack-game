```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

global total
total = 100

print('Welcome to Black Jack! Get as close to 21 as you can without going over!\n\
    Dealer hits until she reaches 17. Aces count as 1 or 11.')

class Card:
    
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
    
    def __str__(self):
        return f'{self.rank} of {self.suit}'
    
class Deck:
    
    def __init__(self):
        self.deck = [] 
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
    
    def __str__(self):
        my_deck = ''
        for card in self.deck:
            my_deck = my_deck + card.__str__() + '\n'
        return my_deck

    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        single_card = self.deck.pop()
        return single_card

class Hand:
    
    def __init__(self):
        self.cards = []  
        self.value = 0  
        self.aces = 0   
    
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank]
        
        if card.rank == 'Ace':
            self.aces += 1
           
    def adjust_for_ace(self):
        if self.aces and self.value > 21:
            self.value -= 10
            self.aces -= 1

class Chips:
    
    def __init__(self, total):
        self.total = total
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
        
def take_bet(chips):
    
    while True:
        try:
            chips.bet = int(input(f'How many chips would you like to bet? You have {chips.total} chips.  '))
            if chips.bet > chips.total:
                print(f'Sorry, you have only {chips.total}.')
                continue
            elif chips.bet < 1:
                print('Sorry, your bet must be over 0.')
                continue
        except:
            print('Sorry, please provide an integer.')
        else:
            break
            
def hit(deck,hand):
    
    hand.add_card(deck.deal())
    hand.adjust_for_ace()
    
def hit_or_stand(deck,hand):
    
    global playing  # to control an upcoming while loop
    
    while True:
        try:
            x = input('\nWould you like to hit or stand? Enter h for HIT or s for STAND.  ')

            if x.lower()[0] == 'h':
                hit(deck,hand)
                show_some(hand,dealer)
            elif x.lower()[0] == 's':
                print("\nPlayer stands.")
                playing = False
            else:
                print("Sorry, I didn't understand that.")
                continue
            break
        except:
            print("Sorry, I didn't understand that.")
            
            
def show_some(player,dealer):
    
    print("\nCards on dealer's hand are:")
    print('<hidden card>')
    print(dealer.cards[1])
    
    print("\nCards on your hand are:")
    for card in player.cards:
        print(card)
        
def show_some_after_split(player,dealer,player_split):
    show_some(player,dealer)
    print("\nCards on your splitted hand are:")
    for card in player_split.cards:
        print(card)
    
def show_all(player,dealer):
    print("\nCards on dealer's hand are:")
    for card in dealer.cards:
        print(card)
    
    print("\nCards on your hand are:")
    for card in player.cards:
        print(card)

def show_all_after_split(player,dealer,player_split):
    show_all(player,dealer)
    print("\nCards on your splitted hand are:")
    for card in player_split.cards:
        print(card)
        
def double_down(chips):
    while True:
        try:
            if_double = input("\nWould you like to double down? After you've doubled your bet you will receive one more card. Enter y for YES or n for NO.  ")
            if if_double.lower()[0] == 'y':
                chips.bet *= 2
                hit(deck,player)
                return False
                break
            elif if_double.lower()[0] == 'n':
                return True
                break
            else:
                print("Sorry, I didn't understand that.")
        except:
            print("Sorry, I didn't understand that.")

def insuring(chips):
    try:
        if_insure = input(f'Would you like to insure the game? Insurane costs {chips.bet} chips and if dealer got Black Jack insurance pays 2:1. ')
        if if_insure.lower()[0] == 'y':
            insurance = True
            chips.total -= chips.bet
    except:
        pass
            
def player_busts(chips):
    print('\nPlayer busts!')
    chips.lose_bet()

def player_wins(chips):
    print('\nPlayer wins!')
    chips.win_bet()

def dealer_busts(chips):
    print('\nDealer busts!')
    chips.win_bet()
    
def dealer_wins(chips):
    print('\nDealer wins!')
    chips.lose_bet()
    
def push():
    print('\nTied game.')

def while_playing(player):
    while playing and player.value < 21: 
        hit_or_stand(deck,player)

        if player.value == 21:
            print("You've got 21 points.")
            break 
            
def win_check(dealer,player,chips):         
    if dealer.value == 21 and len(dealer.cards) == 2:
        print('\nDealer has got Black Jack!')
        dealer_wins(chips)
        if insurance == True:
            chips.total += chips.bet*2
            print(f'You insured the game so you get {chips.bet*2} chips from the insurance.')
    elif dealer.value > 21:
        dealer_busts(chips)
    elif dealer.value < player.value:
        player_wins(chips)
    elif dealer.value > player.value:
        dealer_wins(chips)
    else:
        push()   
    
        
while True:
    split = False
    insurance = False
    black_jack = False
    
    deck = Deck()
    deck.shuffle()
    
    player = Hand()
    dealer = Hand()
    hit(deck, player)
    hit(deck, dealer)
    hit(deck, player)
    hit(deck, dealer)
        
    chips = Chips(total)
    take_bet(chips)    
    show_some(player,dealer)
    
    if player.value == 21:
        print("Con gratulations! You've got Black Jack.")
        show_all(player,dealer)
        playing = False
        black_jack = True
        if dealer.value != 21:
            player_wins(chips)
        else:
            print('\nPlayer and dealer have got Black Jack')
            push()
    else:
        playing = double_down(chips)
    
    if player.cards[0].rank == player.cards[1].rank and (chips.total >= chips.bet) and playing:
        if player.cards[0].rank == 'Ace': print('\nIf you decide to split Aces you will receive one more card on every hand.')
        while True:
            try:
                if_split = input('\nWould you like to split your cards? You will need to bet splitted hand with same number of chips. Enter y for YES, n for NO.  ')
                if if_split.lower()[0] == 'y':
                    split = True
                    player_split = Hand()
                    player_split.cards.append(player.cards.pop())
                    player.value = values[player.cards[0].rank]
                    player_split.value = player.value
                    if player.cards[0].rank == 'Ace':
                        player_split.aces = 1
                    hit(deck,player)
                    hit(deck,player_split)
                    
                    if player.cards[0].rank == 'Ace':
                        while dealer.value < 17:
                            print("\nDealer's cards:")
                            for card in dealer.cards:
                                print(card)
                            print('DELAER HITS')
                            hit(deck,dealer)
                        show_all_after_split(player,dealer,player_split)
                        print('\nOn your first hand:')
                        win_check(dealer,player,chips)
                        print('\nOn your splitted hand:')
                        win_check(dealer,player_split,chips)
                    else:
                        show_some_after_split(player,dealer,player_split)
                        print('\nOn your first hand:')
                        for card in player.cards:
                            print(card)
                        while_playing(player)
                        if player.value > 21:
                            show_some_after_split(player,dealer)
                            player_busts(chips)

                        print('\nOn your splitted hand:')
                        for card in player_split.cards:
                            print(card)
                        playing = True
                        while_playing(player_split)
                        
                        if player_split.value > 21:
                            show_all_after_split(player,dealer,player_split)
                            player_busts(chips)
                        
                        while dealer.value < 17:
                            print("\nDealer's cards:")
                            for card in dealer.cards:
                                print(card)
                            print('DELAER HITS')
                            hit(deck,dealer)
                        
                        if (player.value or player_split.value) <= 21:
                            show_all_after_split(player,dealer,player_split)
                            
                        if player.value <= 21:
                            print('\nOn your first hand:')
                            win_check(dealer,player,chips)
                        if player_split.value <= 21:
                            print('\nOn your splitted hand:')
                            win_check(dealer,player_split,chips)
                    break 
                elif if_split.lower()[0] == 'n':
                    break
                else:
                    print("Sorry, I didn't understand that.") 
            except:
                print("Sorry, I didn't understand that.")

    elif player.cards[0].rank == player.cards[1].rank and chips.total < chips.bet:
            print("You don't have enough chips to split your cards.")
    
    if not split and playing and dealer.cards[1].rank == 'Ace' and chips.total >= chips.bet:
        insuring(chips)
    
    if not split and black_jack == False:
        while_playing(player)
     
    if player.value > 21 and not split:
        show_all(player,dealer)
        player_busts(chips)
        if dealer.value == 21 and len(dealer.cards) == 2 and insurance == True:
            chips.total += chips.bet*2
            print(f"You insured the game so you get {chips.bet*2} chips from the insurance.")
        
    elif black_jack == False and not split:
        
        while dealer.value < 17:
            print("\nDealer's cards:")
            for card in dealer.cards:
                print(card)
            print('DELAER HIT')
            hit(deck,dealer)
    
        show_all(player, dealer)
        win_check(dealer,player,chips)
        
    print(f'\nYou have: {chips.total} chips.')
    total = chips.total
    
    if total > 0:
        try:
            play_again = input('\nWould you like to play again? Enter y for YES ')
            if play_again == 'y':
                continue
            else:
                print('Thank you for the game. Good bye.')
                break
        except:
            print('Thank you for the game. Good bye.')
    else:
        print('You have no more chips. Thank you for the game.')
        break

    
```
